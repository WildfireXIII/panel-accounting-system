```python
import panel as pn
```

```python
css = '''
.bk .mycustomcard {
    background: #119933;
    border-radius: 10px;
    color: white !important;
    margin: 10px;
    line-height: .5;
}


.bk .card-button {
    display: none !important;
}

.bk .curvy {
    border-radius: 20px !important;
}
'''
```

```python
pn.extension('ipywidgets', raw_css=[css])
```

```python
def testcol(name, value):
    return pn.Column(pn.pane.HTML(f"<h1>{name}</h1><p>{value}</p>", margin=[10, 5]), css_classes=['mycustomcard'], margin=5)
        
```

```python
pn.FlexBox(pn.Column(
    pn.pane.HTML("""<h1>Test</h1><p>What?</p>"""),
    css_classes=['mycustomcard']
))
```

```python
from matplotlib.figure import Figure
import numpy as np

fig = Figure(figsize=(1.5, .5))
ax = fig.subplots()

data = [0, 1, 2, 3, 4, 2, 8]
ax.plot(data, color='white', lw=2)
ax.fill_between(x=[0,1,2,3,4,5,6], y1=data, color='white', alpha=0.2)

ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.patch.set_facecolor("orange")
ax.patch.set_alpha(0.0)

fig.patch.set_facecolor("green")
fig.patch.set_alpha(0.0)
ax.margins(x=0, y=0)
fig.subplots_adjust(left=0, bottom=0, right=1, top=1)

pn.pane.Matplotlib(fig, dpi=80, margin=10) 
```

```python
import pandas as pd
df = pd.DataFrame([{"thing1": "yes", "thing2": "no"}, {"thing1": "WHAA", "thing2": "WHAAA??"}])
```

```python
pn.FlexBox(pn.Card(
    pn.pane.DataFrame(df),
    header=pn.FlexBox(pn.Column(
        pn.pane.HTML("<h1>Test</h1><p>What?</p>"),
        pn.pane.Matplotlib(fig, dpi=80, margin=10),
    )),
    background="red",
    header_color="white",
    header_background="red",
    header_css_classes=['curvy'],
    css_classes=['mycustomcard', 'curvy']
))
```

```python
# class ClickableCard
```

```python
mycard = pn.Card(
    pn.pane.DataFrame(df),
    header=pn.FlexBox(pn.Column(
        pn.pane.HTML("<h1>Test</h1><p>What?</p>"),
        pn.pane.Matplotlib(fig, dpi=80, margin=10),
    )),
    background="red",
    header_color="white",
    header_background="red",
    header_css_classes=['curvy'],
    css_classes=['mycustomcard', 'curvy']
)
```

```python
@pn.depends(mycard.param.collapsed, watch=True)
def dothing(things):
    print(things)
    print(mycard.collapsed)

pn.FlexBox(mycard)
```
